```python
import pandas as pd
import matplotlib.pyplot as plt
import torch
%matplotlib inline
```

```python
df = pd.read_csv('ENB2012_data.csv', index_col=None, header=0)
df.columns = ['Relative Compactness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Orientation', 'Glazing Area', 'Glazing Area Distribution', 'Heating Load', 'Cooling Load']
fig, axs = plt.subplots(1, 8, sharey=True) # create plot that will include 8 subplots axs[?]
df.plot(kind='scatter', x='Relative Compactness', y='Cooling Load', ax=axs[0], figsize=(14, 6))
df.plot(kind='scatter', x='Surface Area', y='Cooling Load', ax=axs[1], figsize=(14, 6))
df.plot(kind='scatter', x='Wall Area', y='Cooling Load', ax=axs[2], figsize=(14, 6))
df.plot(kind='scatter', x='Roof Area', y='Cooling Load', ax=axs[3], figsize=(14, 6))
df.plot(kind='scatter', x='Overall Height', y='Cooling Load', ax=axs[4], figsize=(14, 6))
df.plot(kind='scatter', x='Orientation', y='Cooling Load', ax=axs[5], figsize=(14, 6))
df.plot(kind='scatter', x='Glazing Area', y='Cooling Load', ax=axs[6], figsize=(14, 6))
df.plot(kind='scatter', x='Glazing Area Distribution', y='Cooling Load', ax=axs[7], figsize=(14, 6))
```
